#### 2. Storing the data path

```python
#stroing the path of the dataset forlder in the variable
data_path = '20_newsgroups'

#creating a list of folder names to make valid pathnames later
folders = [f for f in listdir(data_path)]
```

#### 4. Importing the requried functions from the sklearn library

```python

import sklearn.neighbors
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score
from sklearn import model_selection
```

```python
files = []
for folder_name in folders:
    folder_path = join(data_path, folder_name)
    files.append([f for f in listdir(folder_path)])
```

#### 6. Taking the path of the each file in the list 

```python
path_list = []
for fo in range(len(folders)):
    for fi in files[fo]:
        path_list.append(join(data_path, join(folders[fo], fi)))
```

#### 7. Print path list

```python
print(path_list)
```

```python
Y = []
for folder_name in folders:
    folder_path = join(data_path, folder_name)
    num_of_files= len(listdir(folder_path))
    for i in range(num_of_files):
        Y.append(folder_name)
```

```python
print(Y)
```

```python

newsdata = fetch_20newsgroups(subset='all', categories=Y)

#initializing the tfidfvectorizer
vectorizer = TfidfVectorizer()

#____ Preprocessing on data using Tf-Idf_____#
X = vectorizer.fit_transform(newsdata.data)
y = newsdata.target

weights = 'uniform' # Using Unifor Disturbuted for weight initialization
  
```

#### 8. Defining the classifier(KNN)

```python

clf = sklearn.neighbors.KNeighborsClassifier(n_neighbors=1, weights=weights)
```

#### 9. K-Fold (5-Fold)

```python

def validation(data, target, classifier, cv=5):
    return sklearn.model_selection.cross_val_score(classifier, data, target, cv=cv)
```

#### 10. Test classifier

```python
def test_classifier(X, y, clf, test_size=0.4, y_names=None):
    # spliting the dataset into the train and test format
    print('test size is: %2.0f%%' % (test_size * 100))
    X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=test_size)

    #training the model
    clf.fit(X_train, y_train)
    #predicting the valute of test dataset
    y_predicted = clf.predict(X_test)

    #printing the accuracy score
    print ("accuracy of the model before checking cross-validation is %0.2f" % accuracy_score(y_test, y_predicted))

    #checking the cross validation
    scores = validation(X, y, clf, cv=5)
    print("Accuracy of the model after cross-validation: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))

    #generating the classification report
    print('Classification report:')
    print(sklearn.metrics.classification_report(y_test, y_predicted, target_names=y_names))
```

#### 11. Function call

```python
#calling the function
test_classifier(X,y,clf) 
```
